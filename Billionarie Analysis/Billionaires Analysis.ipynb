```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Billionaire.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
# since networth has object datatype we will be changing it into float and cleaning it with removing '$' sign and 'B'

df['NetWorth'] = df['NetWorth'].str.strip("$")
df['NetWorth'] = df['NetWorth'].str.strip("B")
df['NetWorth'] = df['NetWorth'].astype(float)
```

```python
# checking null values
print(df.isnull().sum())
```

```python
# we will be dropping the age
df.info()
```

```python
# we will be dropping the null values since the count of null values is less compared to overall dataframe size , 
#it won't affect the final observation a lot
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Top 10 billionaries according to the networth

df2 = df.sort_values(by = ["NetWorth"], ascending=False).head(10)
plt.figure(figsize=(20, 10))
sns.histplot(x="Name", hue="NetWorth", data=df2)
plt.show()
```

```python
#Now lets have a look at the top 5 domains with the most number of billionaires:

a = df["Source"].value_counts().head()
index = a.index
sources = a.values
custom_colors = ["skyblue", "yellowgreen", 'tomato', "blue", "red"]
plt.figure(figsize=(5, 5))
plt.pie(sources, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()           #gcf function gives the current figure
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Domains to Become a Billionaire", fontsize=20)
plt.show()
```

```python
#Now lets have a look at the top 5 industries with the most number of billionaires:

a = df["Industry"].value_counts().head()
index = a.index
industries = a.values
custom_colors = ["skyblue", "yellowgreen", 'tomato', "blue", "red"]
plt.figure(figsize=(5, 5))
plt.pie(industries, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Industries with Most Number of Billionaires", fontsize=20)
plt.show()
```

```python
#Now lets have a look at the top 5 countries with the most number of billionaires:

a = df["Country"].value_counts().head()
index = a.index
Countries = a.values
custom_colors = ["skyblue", "yellowgreen", 'tomato', "blue", "red"]
plt.figure(figsize=(5, 5))
plt.pie(Countries, labels=index, colors=custom_colors)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Top 5 Countries with Most Number of Billionaires", fontsize=20)
plt.show()
```
